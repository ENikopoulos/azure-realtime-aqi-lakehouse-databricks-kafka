```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
spark = SparkSession.builder.appName("TestWriteToLakehouse").getOrCreate()
```

```python
test_df = spark.createDataFrame(
    [("abc", 123)],
    ["name", "value"]
)

output_path = "abfss://lakehouse@staqilakehouseen.dfs.core.windows.net/bronze/test/"

test_df.write.mode("overwrite").parquet(output_path)

spark.read.parquet(output_path).show()
```
